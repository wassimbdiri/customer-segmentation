```python
df.loc[df['Gender'] == 'Male', 'Gender'] = 1
df.loc[df['Gender'] == 'Female', 'Gender'] = 0
```

```python
df.loc[df['Ever_Married'] == 'Yes', 'Ever_Married'] = 1
df.loc[df['Ever_Married'] == 'No', 'Ever_Married'] = 0
```

```python
df.loc[df['Graduated'] == 'Yes', 'Graduated'] = 1
df.loc[df['Graduated'] == 'No', 'Graduated'] = 0
```

```python
df.loc[df['Segmentation'] == 'A', 'Segmentation'] = 0
df.loc[df['Segmentation'] == 'B', 'Segmentation'] = 1
df.loc[df['Segmentation'] == 'C', 'Segmentation'] = 2
df.loc[df['Segmentation'] == 'D', 'Segmentation'] = 3
```

```python
df.loc[df['Spending_Score'] == 'Low', 'Spending_Score'] = 1
df.loc[df['Spending_Score'] == 'Average', 'Spending_Score'] = 2
df.loc[df['Spending_Score'] == 'High', 'Spending_Score'] = 3
```

```python
df.loc[df['Var_1'] == 'Cat_1', 'Var_1'] = 1
df.loc[df['Var_1'] == 'Cat_2', 'Var_1'] = 2
df.loc[df['Var_1'] == 'Cat_3', 'Var_1'] = 3
df.loc[df['Var_1'] == 'Cat_4', 'Var_1'] = 4
df.loc[df['Var_1'] == 'Cat_5', 'Var_1'] = 5
df.loc[df['Var_1'] == 'Cat_6', 'Var_1'] = 6
df.loc[df['Var_1'] == 'Cat_7', 'Var_1'] = 7
```

```python
df=df.drop('ID',axis=1)
```

```python
df.head()
```

```python
df=df.fillna(df.mode())
```

```python
df
```

```python
values = {'Ever_Married': df["Ever_Married"].mean(), 'Graduated': df["Graduated"].mean(), 'Profession': "Notfound" , 
          'Work_Experience': df["Work_Experience"].mean(), 'Family_Size': df['Family_Size'].mean(), 'Var_1':df["Var_1"].mean() }
df=df.fillna(value=values)
```

```python
y=df["Segmentation"]
x=df.drop("Segmentation",axis=1)
y=y.astype('int') 
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
enc = OneHotEncoder(handle_unknown='ignore')
enc_df=pd.DataFrame(enc.fit_transform(x[['Profession']]).toarray())
x= x.join(enc_df)
#x.merge(x, how='left')
x=x.drop('Profession',axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
model1 = RandomForestClassifier(max_depth=2, random_state=0)
model1.fit(x_train, y_train)
y_pred=model1.predict(x_test)
y_predtrain=model1.predict(x_train)
print('the accuracy for the test data is',accuracy_score(y_test, y_pred))
print('the accuracy for the train data is',accuracy_score(y_train, y_predtrain))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() 
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model4 = KNeighborsClassifier(n_neighbors=9)
model4.fit(x_train, y_train)
y_pred=model4.predict(x_test)
y_predtrain=model4.predict(x_train)
print('the accuracy for the test data is',accuracy_score(y_test, y_pred))
print('the accuracy for the train data is',accuracy_score(y_train, y_predtrain))
```

```python
from sklearn import svm
model = svm.SVC()
model.fit(x_train ,y_train)
y_pred=model.predict(x_test)
y_predtrain=model.predict(x_train)
print('the accuracy for the test data is',accuracy_score(y_test, y_pred))
print('the accuracy for the train data is',accuracy_score(y_train, y_predtrain))
```

```python
import tensorflow as tf 
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(18,)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(4)
])
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
model.fit(x_train,y_train, epochs=30)
test_loss, test_acc = model.evaluate(x_test,  y_test, verbose=2)

print('\nTest accuracy:', test_acc)
```
